```python
%load_ext autoreload
%autoreload 1


from PIL import Image
import numpy as np
from src import detect_faces, show_bboxes, get_list_of_files, resize_with_pad
import matplotlib.pyplot as plt
import torch
import torchvision.transforms.functional as F
import os

from src import PNet, RNet, ONet
from random import shuffle




%aimport src
```

```python
face_list = get_list_of_files('./test/')
shuffle(face_list)
max_images = 100
face_list = face_list[:max_images]
```

```python
count = 0
for filename in face_list:
    image = Image.open(filename)
    bounding_boxes, landmarks = detect_faces(image)
    if isinstance(bounding_boxes, np.ndarray):
        if bounding_boxes.shape[0] != 0:
            image.crop((bounding_boxes[0][0], bounding_boxes[0][1]-100, \
                     bounding_boxes[0][2], bounding_boxes[0][3]+100)).save('./cropped_bigger/cropped_'+str(count)+'.jpg')
            count = count + 1
```

```python
show_bboxes(image, bounding_boxes, landmarks)
face_num = 4
image.crop((bounding_boxes[face_num][0], bounding_boxes[face_num][1], 
                bounding_boxes[face_num][2], bounding_boxes[face_num][3])).save('../crop_0005_01.jpg')
```
